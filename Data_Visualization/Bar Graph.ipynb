```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
product_category = np.array(['Furniture', 'Technology', 'Office Supplies'])
sales = np.array([4110451.90, 4744557.50, 3787492.52])
```

```python
product_category ,sales
```

```python
plt.bar(product_category, sales)
plt.show()
```

```python
plt.xlabel('Procuct Category',fontdict={'fontsize':12, 'fontweight':5,'color':'gray'})
plt.ylabel('Sales',fontdict={'fontsize':12, 'fontweight':5, 'color':'gray'})
plt.bar(product_category, sales, width = 0.5, color = 'pink', edgecolor = 'gray', align = 'center')
plt.title('Sales per Product Category', color = "gray", fontdict={'fontsize':14, 'fontweight':5})

plt.show()
```
